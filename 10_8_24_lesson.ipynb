# Pearson correlation

The correlation coefficient indicates the strength and direction of the statistical relationship between two variables. The coefficient ranges from -1 to 1, with a value of 0.43 signifying a positive but moderately strong relationship between the two variables.

- A positive correlation coefficient means that the variables increase together. Therefore, as one variable increases, the other also tends to increase, and vice versa.

The closer the correlation coefficient is to 1, the stronger the positive relationship.

- A correlation coefficient of 0.43 is not very high, but it indicates a moderately positive relationship.

If the correlation coefficient were negative, it would suggest that as one variable increases, the other variable decreases.
- A correlation coefficient of -1 indicates a perfect negative relationship.

Thus, a correlation coefficient of 0.43 might suggest that, for example, as Pok√©mon's Attack values increase, their Defense values tend to increase as well, but the relationship is not very strong.

```python
# result of this correlation calculation is between -1 and 1.
# -1 perfect negative relationship
# 1 perfect positive relationship
```

```python
# select only numeric colums

numeric_colums = pokemon_df.select_dtypes(include=['number'])
numeric_colums = numeric_colums.drop('#', axis=1)

numeric_colums
```

```python
target_column  = 'Attack'
correlation_result = {} #key = column, val =pearson_correlation

# for loop comparing target column and all other columns  in the Dataframe

for col in numeric_colums.columns:
    correlation = pokemon_df[target_column].corr(pokemon_df[col])
    correlation_result[col] = correlation

correlation_result
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
pokemon_df = pd.read_csv('/content/Pokemon.csv')

# Create DataFrames for Grass and Water Type 1 Pok√©mon
grass_df = pokemon_df[pokemon_df['Type 1'] == 'Grass']
water_df = pokemon_df[pokemon_df['Type 1'] == 'Water']

# Regression plot for Grass Type 1 Pok√©mon
plt.figure(figsize=(10, 6))
sns.regplot(x='Attack', y='Defense', data=grass_df, color='green')
plt.title('Attack vs. Defense for Grass Type 1 Pok√©mon')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()

# Regression plot for Water Type 1 Pok√©mon
plt.figure(figsize=(10, 6))
sns.regplot(x='Attack', y='Defense', data=water_df, color='blue')
plt.title('Attack vs. Defense for Water Type 1 Pok√©mon')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()

# Calculate Pearson correlation for Grass Type 1 Pok√©mon
grass_correlation = grass_df['Attack'].corr(grass_df['Defense'])
print(f'Pearson correlation for Grass Type 1 Pok√©mon (Attack vs. Defense): {grass_correlation:.2f}')

# Calculate Pearson correlation for Water Type 1 Pok√©mon
water_correlation = water_df['Attack'].corr(water_df['Defense'])
print(f'Pearson correlation for Water Type 1 Pok√©mon (Attack vs. Defense): {water_correlation:.2f}')
```

# Explanation
1. Creating DataFrames for Grass and Water Type 1 Pok√©mon:
- We first filter the original pokemon_df DataFrame to create two new DataFrames: grass_df for Grass Type 1 Pok√©mon and water_df for Water Type 1 Pok√©mon.

- grass_df contains all rows where the 'Type 1' column is 'Grass'.
- water_df contains all rows where the 'Type 1' column is 'Water'

2. Regression Plots:
- We use Seaborn's regplot function to create scatter plots with regression lines to visualize the relationship between Attack and Defense variables for both Grass and Water Type 1 Pok√©mon.
- The first plot shows the relationship for Grass Type 1 Pok√©mon, with points and the regression line in green.
- The second plot shows the relationship for Water Type 1 Pok√©mon, with points and the regression line in blue.
3. Calculating Pearson Correlation:
- We calculate the Pearson correlation coefficient for the Attack and Defense variables in both grass_df and water_df.
- The Pearson correlation coefficient is a measure of the linear relationship between two variables, ranging from -1 to 1.
  - A positive value indicates a positive relationship: as one variable increases, the other also tends to increase.
  - A negative value indicates a negative relationship: as one variable increases, the other tends to decrease.
  - A value close to 0 indicates a weak or no linear relationship.
4. Results:
- For Grass Type 1 Pok√©mon, the Pearson correlation coefficient between Attack and Defense is printed.
- For Water Type 1 Pok√©mon, the Pearson correlation coefficient between Attack and Defense is also printed.
