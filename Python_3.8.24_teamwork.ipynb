To complete the tasks use 'pokemon.csv' dataset Answer the following questions:
1. How many Pokémons are with 'Type 1' == Water as a % of total?
2. What is the maximum 'Speed' value? What is the minimum 'Speed' value? What is the difference between max and min 'Speed'?
3. Filter the DataFrame to include only the Pokémon with 'Speed' >= 80. How many Pokémon meet this criterion? Display this DataFrame using your preferred visualization method.

```python
# Girta:

import pandas as pd
import matplotlib.pyplot as plt
pokemon_df = pd.read_csv('/content/Pokemon.csv')
```

```python
# 1. how many Pokémons are with 'Type 1' == Water as a % of total?

total_pokemons = len(pokemon_df)
water_pokemons = len(pokemon_df[pokemon_df['Type 1'] == 'Water'])
water_pokemon_percentage = (water_pokemons / total_pokemons) * 100
print(f'Percentage of Water Type 1 Pokemons: {water_pokemon_percentage:.2f}%')
```

```python
# 2. what is the maximum 'Speed' value? what is the minimum 'Speed' value? what is the difference between max and min 'Speed'?

max_speed = pokemon_df['Speed'].max()
min_speed = pokemon_df['Speed'].min()
speed_difference = max_speed - min_speed
print(f'Maximum Speed: {max_speed}')
print(f'Minimum Speed: {min_speed}')
print(f'Difference between max and min Speed: {speed_difference}')
```

```python
# 3. filter the DataFrame to include only the Pokémon with 'Speed' >= 80. how many Pokémon meet this criterion? display this DataFrame using your preferred visualization method

fast_pokemons = pokemon_df[pokemon_df['Speed'] >= 80]
print(f'Number of Pokemons with Speed >= 80: {len(fast_pokemons)}')
print(fast_pokemons.head())

# display filtered DataFrame

plt.figure(figsize=(12, 6))
plt.hist(fast_pokemons['Speed'], bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Pokemon Speeds >= 80')
plt.xlabel('Speed')
plt.ylabel('Number of Pokemons')
plt.show()
```

```python
# Katrin:

import pandas as pd #analysis
import matplotlib.pyplot as plt # vizualization
pokemon_df = pd.read_csv('/content/Pokemon.csv')
```

```python
#1 Calculate the number of Pokémon with 'Type 1' == 'Water'
water_pokemon_count = pokemon_df[pokemon_df['Type 1'] == 'Water'].shape[0]

total_pokemon_count = pokemon_df.shape[0]

water_pokemon_percentage = (water_pokemon_count / total_pokemon_count) * 100

print(f"Percentage of Pokémon with 'Type 1' == 'Water': {water_pokemon_percentage:.2f}%")

#2 Maximum and minimum speed; speed difference
max_speed = pokemon_df['Speed'].max()
min_speed = pokemon_df['Speed'].min()
speed_difference = max_speed - min_speed

print(f"Maximum Speed: {max_speed}")
print(f"Minimum Speed: {min_speed}")
print(f"Difference between max and min Speed: {speed_difference}")

#3 Pokémon with 'Speed' >= 80
speed_pokemon = pokemon_df[pokemon_df['Speed'] >= 80]
speed_count = speed_pokemon.shape[0]
print(f'Number of Pokémon with Speed >= 80: {speed_count}')

# Vizualize this DataFrame with histogram
speed_high_80 = pokemon_df[pokemon_df['Speed'] >= 80]
print(f'{len(speed_high_80)} Speed is more than 80')
plt.hist(pokemon_df['Speed'], bins=30, color='green', edgecolor='black')
plt.xlabel('Speed')
plt.ylabel('Number of pokemons (Frequency)')

#4 Pokemon with the longest name
pokemon_df['Name_Length'] = pokemon_df['Name'].str.replace(' ', '').str.len()

longest_name_pokemon = pokemon_df.loc[pokemon_df['Name_Length'].idxmax(), 'Name']

print(f'The Pokémon with the longest name (excluding spaces) is: {longest_name_pokemon}')

```
