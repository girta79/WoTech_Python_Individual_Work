```python
# data object

# Sample data representing the average temperature (in degrees Celsius) from 2010 to 2014
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Avg_Temperature': [14.1, 14.3, 14.5, 14.2, 14.6]}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
df
```

```python
# VERSION ONE

plt.plot(df['Year'], df['Avg_Temperature'], marker="o") # first value is x, second y
plt.title("Average temperature from 2010 to 2014") # we add title
plt.xlabel("Year") # title on x axis
plt.ylabel("Temperature") # title on y axis

plt.xticks(df['Year']) # we add ticks on x axis (year values are rounded)
plt.grid(True) # we add grid
plt.show() # show the visualization
```

```python
# VERSION TWO

df.plot(x='Year', y='Avg_Temperature', kind='line') # we are using pandas
plt.title('Average temperature from 2010 to 2014') # add title
plt.xlabel('Year') # title on x axis
plt.ylabel('Temperature') # title on y axis

plt.xticks(df['Year']) # year values are rounded
plt.grid(True)
plt.show() # show the visualization
```

The difference between df.plot() and plt.plot() lies in the library-specific functions used and the level of control and customization each offers. Here are the details:

# **df.plot()**


This is a high-level plotting function provided by the pandas library, which acts as a wrapper around matplotlib. It simplifies the process of creating a plot directly from a pandas DataFrame. (When we say that df.plot() in pandas is a "wrapper around matplotlib," it means that the pandas plotting functions are built on top of matplotlib.)

**Pros:**

1. **Simplicity:** Easy to use, especially for quick plots directly from a DataFrame.
2. **Integration:** Works directly with DataFrame columns, making it very convenient for exploratory data analysis.
3. **Less Code:** Requires less code to generate a plot.

**Cons:**

1. **Less Customization:** While it's great for quick plots, it offers less fine-grained control over the plot appearance compared to matplotlib.


# **plt.plot()**

This is a low-level plotting function from the matplotlib library. It requires more code but offers extensive customization options for creating detailed and complex plots.

**Pros:**

1. **Customization:** Allows for extensive customization of all aspects of the plot (titles, labels, colors, markers, grid, etc.).
2. **Flexibility:** Suitable for creating complex and highly customized plots.
3. **Control:** Provides more control over the plotting process and appearance.

**Cons:**

1. **Complexity**: Requires more lines of code and a deeper understanding of matplotlib.


**Summary**

**df.plot()** is convenient for quick, simple plots directly from a DataFrame with minimal code.
**plt.plot()** offers more detailed and customizable plotting options but requires more code and understanding of the matplotlib library.

Both methods are useful depending on the context: use df.plot() for quick visualization and plt.plot() for more detailed and customized plots. Code Abstraction By abstracting the complexity of matplotlib, pandas plotting functions make it easier to create standard plots quickly. For users who need basic visualizations, this means they can focus on their data rather than learning the intricacies of matplotlib. Customization with Matplotlib While pandas provides a simplified interface, the plots created are still matplotlib objects. This means you can further customize these plots using matplotlib functions if needed.

```python
# SCATTER PLOT - used to visualize the relationship between two variables

data3 = {'X': [1,2,3,5,5,2], 'Y':[5,8,7,4,9,5]}
df3 = pd.DataFrame(data3)

plt.scatter(x = df3['X'], y=df3['Y'])
plt.title('Scatter Plot')
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# task 1.

import pandas as pd
import matplotlib.pyplot as plt

file_name = 'transaction_dataset.csv'
df = pd.read_csv(file_name)

print(df.head())

gender_counts = df['Gender'].value_counts(dropna=False)

gender_counts.plot(kind='bar')
plt.title('Count of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# task 2.

import pandas as pd
import matplotlib.pyplot as plt

file_name = 'transaction_dataset.csv'
df = pd.read_csv(file_name)

print(df.head())

name_counts = df['Name'].value_counts().reset_index()
name_counts.columns = ['Name', 'Count']

top_5_names = name_counts.iloc[:5]

plt.figure(figsize=(10, 6))
plt.barh(top_5_names['Name'], top_5_names['Count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('Name')
plt.title('Top 5 Most Frequent Names')
plt.gca().invert_yaxis()  # AI help: invert the Y-axis so that the most frequent name is at the top
plt.show()
```

```python
# task 3. AI:

import pandas as pd

file_name = 'transaction_dataset.csv'
df = pd.read_csv(file_name)

print(df.head())

# filter the DataFrame where Category is 'Clothing' and Gender is 'M'
filtered_df = df[(df['Category'] == 'Clothing') & (df['Gender'] == 'M')]

# get the number of rows in the filtered DataFrame
num_rows = len(filtered_df)

# output the result
result_message = f"The filtered DataFrame has {num_rows} rows"
print(result_message)
```
